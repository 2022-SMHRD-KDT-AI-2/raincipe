```python
# 필요한 라이브러리 임포트
import os
import re
from google.cloud import texttospeech_v1
from google.cloud import speech_v1p1beta1 as speech
from flask import Flask
from flask import request, redirect, render_template
from flask_cors import CORS
import playsound
import pyaudio
import wave
from datetime import datetime
```

```python
def convert(word):
    times = {'초': 1, '분': 60}
    if "다음" in word:
        return "다음"
    elif "이전" in word:
        return "이전"
    elif "다시" in word:
        return "다시"
    elif '초' in word or '분' in word:
        word_lst = re.findall('\d+\D', word)
        time = 0
        for i in word_lst:
            lst = list(filter(None, re.split('(\d+)', i)))
            time += int(lst[0]) * int(times[lst[1]])
        return str(time)
    else:
        return "재녹음"
```

```python
def recording():
    # 현재 시간 체크
    time = datetime.now()
    # 녹음을 위한 pyaudio 셋업
    audio = pyaudio.PyAudio()
    # audio config 세팅
    stream = audio.open(format=pyaudio.paInt16, channels=1, rate=44100, input=True, frames_per_buffer=1024)
    # 녹음들을 저장하기 위한 리스트
    frames = []
    # 5초간 녹음
    while (datetime.now() - time).seconds <= 3:
        data = stream.read(1024)
        frames.append(data)
    # 5초후 스트리밍 종료
    stream.stop_stream()
    stream.close()
    audio.terminate()
    # 오디오 저장
    sound_file = wave.open("my_recording.wav", "wb")
    sound_file.setnchannels(1)
    sound_file.setsampwidth(audio.get_sample_size(pyaudio.paInt16))
    sound_file.setframerate(44100)
    sound_file.writeframes(b''.join(frames))
    sound_file.close()
```
