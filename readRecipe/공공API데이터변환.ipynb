```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
import pandas as pd
```

```python
url= "http://openapi.foodsafetykorea.go.kr/api/937d02a68574482897c0/COOKRCP01/xml/1/1000"
result = requests.get(url)
soup = BeautifulSoup(result.text,"lxml")
print(soup)
```

```python
menus=[]
images=[]
classifications=[]
cooks=[]
ingredients=[]
manual01s=[]
manual02s=[]
manual03s=[]
manual04s=[]
manual05s=[]
manual06s=[]
manual07s=[]
manual08s=[]
manual09s=[]
manual10s=[]
manual11s=[]
manual12s=[]
manual13s=[]
manual14s=[]
manual15s=[]
manual16s=[]
manual17s=[]
manual18s=[]
manual19s=[]
manual20s=[]

length = soup.find_all('row')

for i in range(len(length)):
    menu=soup.select('rcp_nm')
    image=soup.select('ATT_FILE_NO_MK')
    classification=soup.select('rcp_pat2')
    cook=soup.select('rcp_way2')
    ingredient=soup.select('rcp_parts_dtls')
    manual01=soup.select('manual01')
    manual02=soup.select('manual02')
    manual03=soup.select('manual03')
    manual04=soup.select('manual04')
    manual05=soup.select('manual05')
    manual06=soup.select('manual06')
    manual07=soup.select('manual07')
    manual08=soup.select('manual08')
    manual09=soup.select('manual09')
    manual10=soup.select('manual10')
    manual11=soup.select('manual11')
    manual12=soup.select('manual12')
    manual13=soup.select('manual13')
    manual14=soup.select('manual14')
    manual15=soup.select('manual15')
    manual16=soup.select('manual16')
    manual17=soup.select('manual17')
    manual18=soup.select('manual18')
    manual19=soup.select('manual19')
    manual20=soup.select('manual20')
    
    menus.append(menu[i].text.replace('\r\n' , ' '))
    images.append(image[i].text)
    classifications.append(classification[i].text.replace('\r\n' , ' '))
    cooks.append(cook[i].text.replace('\r\n' , ' '))
    ingredients.append(ingredient[i].text.replace('\r\n' , ' '))
    manual01s.append(manual01[i].text.replace('a', '').replace('b','').replace('c',''))
    manual02s.append(manual02[i].text.replace('a', '').replace('b','').replace('c',''))
    manual03s.append(manual03[i].text.replace('a', '').replace('b','').replace('c',''))
    manual04s.append(manual04[i].text.replace('a', '').replace('b','').replace('c',''))
    manual05s.append(manual05[i].text.replace('a', '').replace('b','').replace('c',''))
    manual06s.append(manual06[i].text.replace('a', '').replace('b','').replace('c',''))
    manual07s.append(manual07[i].text.replace('a', '').replace('b','').replace('c',''))
    manual08s.append(manual08[i].text.replace('a', '').replace('b','').replace('c',''))
    manual09s.append(manual09[i].text.replace('a', '').replace('b','').replace('c',''))
    manual10s.append(manual10[i].text.replace('a', '').replace('b','').replace('c',''))
    manual11s.append(manual11[i].text.replace('a', '').replace('b','').replace('c',''))
    manual12s.append(manual12[i].text.replace('a', '').replace('b','').replace('c',''))
    manual13s.append(manual13[i].text.replace('a', '').replace('b','').replace('c',''))
    manual14s.append(manual14[i].text.replace('a', '').replace('b','').replace('c',''))
    manual15s.append(manual15[i].text.replace('a', '').replace('b','').replace('c',''))
    manual16s.append(manual16[i].text.replace('a', '').replace('b','').replace('c',''))
    manual17s.append(manual17[i].text.replace('a', '').replace('b','').replace('c',''))
    manual18s.append(manual18[i].text.replace('a', '').replace('b','').replace('c',''))
    manual19s.append(manual19[i].text.replace('a', '').replace('b','').replace('c',''))
    manual20s.append(manual20[i].text.replace('a', '').replace('b','').replace('c',''))
```

```python
recipe=pd.DataFrame({'요리명':menus, '요리사진':images, '분류':classifications,'요리 방법':cooks, '재료':ingredients,
                   '순서1':manual01s,'순서2':manual02s,'순서3':manual03s,'순서4':manual04s,'순서5':manual05s,
                   '순서6':manual06s,'순서7':manual07s,'순서8':manual08s,'순서9':manual09s,'순서10':manual10s,
                   '순서11':manual11s,'순서12':manual12s,'순서13':manual13s,'순서14':manual14s,'순서15':manual15s,
                   '순서16':manual16s,'순서17':manual17s,'순서18':manual18s,'순서19':manual19s,'순서20':manual20s})
```
